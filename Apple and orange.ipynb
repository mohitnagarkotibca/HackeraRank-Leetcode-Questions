```python
s = 7
t = 10
a = 4
b = 12
m = 3
n = 2

apples = [2,3,-4]
oranges = [3,-2,-4]

def countApplesAndOranges(s, t, a, b, apples, oranges):
    fall_apple= [x+a for x in apples]
    fall_orange= [x+b for x in oranges]
    apples= [apple for apple in fall_apple if apple in range(s,t+1)]
    oranges= [orange for orange in fall_orange if orange in range(s,t+1)]   
    print(len(apples))
    print(len(oranges))
countApplesAndOranges(s, t, a, b, apples, oranges)
```
