```python
def pageCount(n, p):
    #foward
    if n==p:
        return 0
    j=0
    for i in range(1,n+1,2):
        if p <= i:
            left= j
            break
        j+=1
    
    if n%2==0:
        limit= int(n/2)
        right= limit-left
    else:
        limit=int((n+1)/2)
        right= limit-left-1
    if left < right:
        return left
    else:
        return right


print(pageCount(9,7))
print('---')
pageCount(10,7)
```
