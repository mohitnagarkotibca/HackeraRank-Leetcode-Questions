```python
def dayOfProgrammer(year):
    if year == 1918:
        return '26.09.1918'
    if year >= 1918:
        t='G'
    else:
        t='J'
        
    if t=='G':
        if ((year%400 ==0) |(year%4==0) & (year%100!=0)):
            return '12.09.'+str(year)
        else:
            return '13.09.'+str(year)
    else:
        if (year%4==0):
            return '12.09.'+str(year)
        else:
            return '13.09.'+str(year)
    
dayOfProgrammer(1800)
```
