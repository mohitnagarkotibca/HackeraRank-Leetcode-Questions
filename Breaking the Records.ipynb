```python
def breakingRecords(scores):
    minimum= scores[0]
    maximum= scores[0]
    max_count=0
    min_count=0
    
    for i in range(1,len(scores)):
        if scores[i]>maximum:
            max_count+=1
            maximum= scores[i]
            
        if scores[i]< minimum:
            min_count+=1
            minimum= scores[i]
            
    return max_count,min_count

scores=[12, 62, 77, 95, 71, 84, 78, 95, 95, 68]
breakingRecords(scores)
```
